# Pathfinding

In railway and airline networks, **finding the shortest path** is critical for efficient travel and transportation planning. Let's explain this from the perspective of graph theory.

First, each station or airport is modeled as a node (vertex), and the railway or airline routes connecting them are modeled as edges. In such networks, finding the shortest path from the starting point to the destination can help reduce time and cost.

For example, by finding the shortest path from City A to City B, a traveler can reach their destination in the shortest possible time. Similarly, logistics companies can minimize delivery costs and time by using the shortest routes. **Various algorithms in graph theory are used to solve such shortest path problems.**

One of the most common algorithms is **Breadth-First Search (BFS)**. BFS is suitable for finding the shortest path in undirected or unweighted graphs. Specifically, it starts at the departure point, explores nodes in a broad manner, and finds the path to the destination in the fewest steps.

Another important method is **Depth-First Search (DFS)**. DFS explores as deeply as possible and then **backtracks** (goes back one step when it hits a dead end) to explore other routes. DFS is useful when you need to confirm the existence of a specific path.

Additionally, **Dijkstra's Algorithm** is one of the most widely used methods for solving shortest path problems. Dijkstra’s algorithm works on **weighted graphs** and finds the shortest path from the starting point to all other nodes. **It progresses by updating the shortest distance to each node, selecting the closest unvisited node for exploration.** This allows for selecting efficient routes, saving fuel costs, and shortening flight times.

In this way, finding the shortest path in railway and airline networks is crucial for travelers and logistics companies. By using algorithms such as Breadth-First Search, Depth-First Search, and Dijkstra’s Algorithm from graph theory, optimal routes can be effectively identified. Different algorithms are used depending on the structure of the graph and the specific pathfinding objectives.

### 1. Breadth-First Search (BFS)
- **Feature**: Explores nodes in order of proximity to the starting vertex.
- **Use case**: Finding the shortest path, suitable for undirected or unweighted graphs.
- **Algorithm**:
  1. Add the starting vertex to the queue and mark it as visited.
  2. While the queue is not empty, remove the first vertex from the queue, and add its unvisited adjacent vertices to the queue.

### 2. Depth-First Search (DFS)
- **Feature**: Explores as deeply as possible.
- **Use case**: Detecting connected components, topological sorting, detecting cycles.
- **Algorithm**:
  1. Add the starting vertex to the stack and mark it as visited.
  2. While the stack is not empty, remove the top vertex from the stack, and add its unvisited adjacent vertices to the stack.

### 3. Dijkstra’s Algorithm
- **Feature**: Finds the shortest path in **non-negative weighted graphs**.
- **Use case**: Shortest pathfinding in weighted graphs.
- **Algorithm**:
  1. Initialize the distance from the starting vertex, and use a priority queue to explore the shortest paths.

### 4. Bellman-Ford Algorithm
- **Feature**: Finds the shortest path even in graphs with **negative weights**.
- **Use case**: When there are edges with negative weights.
- **Algorithm**:
  1. Perform a relaxation operation on all edges in the graph repeatedly to find the shortest path.

### 5. Max-Flow Min-Cut Theorem
- **Feature**: Describes that the **maximum flow** in a network is equal to the **capacity of the minimum cut** of the network.
- **Use case**: Used in fields like traffic systems, packet routing, and resource allocation.
- **Algorithm**:
  1. Use algorithms such as the Ford-Fulkerson method or the Edmonds-Karp algorithm to iteratively find augmenting paths and maximize the flow.
  - **Ford-Fulkerson method**: Finds augmenting paths using residual graphs and increases the flow iteratively.
  - **Edmonds-Karp method**: Efficiently finds augmenting paths using BFS to find the maximum flow.

### 6. A* Algorithm
- **Feature**: Efficiently finds the shortest path using a **heuristic function**.
- **Use case**: Commonly used in game development and map applications for shortest pathfinding.
- **Algorithm**:
  1. Explore paths that have the smallest sum of the actual distance and the estimated distance.

By utilizing these algorithms, you can effectively find the optimal routes in networks, whether it's for travel, transportation, or even gaming applications.

## Breadth-First Search (BFS)

**Breadth-First Search (BFS)** is one of the most intuitive and easy-to-understand graph or tree traversal algorithms. BFS starts from a given starting point (source) and first explores **all adjacent nodes (nodes directly connected to the source)**. Then, it proceeds to explore the adjacent nodes of those adjacent nodes, and so on, gradually expanding the search area level by level.

Consider the example of a maze. Imagine you are standing at the entrance of the maze. First, you would take one step along every possible path directly from the entrance. Then, at each new location, you would take another step along each path you can take from there. This way, instead of following one path all the way to the end, you systematically explore all paths equally, step by step.

Another relatable example is your friend list on a social networking site (SNS). Think of your direct friends as the first level, your friends' friends as the second level, and their friends as the third level, and so on. BFS starts by exploring all your direct friends first, then moves to explore all of your friends' friends, and continues to expand from there.

The specific steps of BFS are as follows:

1. **Add the source node to a queue**. A queue is a data structure where the first element added is the first one removed (FIFO: First In, First Out).
2. **As long as the queue is not empty**, repeat the following process:
   - Remove a node from the front of the queue and mark it as "visited."
   - Add all unvisited adjacent nodes of the current node to the queue.

This process ensures that BFS explores level by level, expanding outwards as it progresses.

The algorithm follows a strategy of "exploring all nearby nodes first." In other words, it visits all reachable unvisited nodes from the current node, and then moves on to visit the unvisited nodes connected to those. This is why it’s called "breadth-first" search.

In this way, **Breadth-First Search (BFS)** is intuitive and straightforward, playing a key role in various fields of study.

## Visualization Based on BFS Distance

**Visualization based on BFS distance** helps to make the operation of the algorithm easier to understand.

By executing this script, you can visually understand the results of the breadth-first search (BFS) on the Karate Club graph **by color-coding according to the distance.**

**1. Generation of the Karate Club Graph**  
- The Zachary Karate Club graph is generated.

**2. Initialization Settings**  
- The lists for node visit status (`known`), distance from the starting node (`dist`), color based on distance (`colors`), and node color management (`color_map`) are initialized.  
- The distance of the starting node (here, node 0) is set to 0, and the color of that node is assigned. The starting node is also marked as visited.

**3. Execution of Breadth-First Search (BFS)**  
- The BFS explores all nodes in the graph and calculates the distance from the starting node for each one.  
- Based on the calculated distance, the color of each node is updated. Different distances are mapped to different colors (red, gray, light gray, white).

**4. Display of Results**  
- The distance of each node is printed to the console, allowing you to verify the distance from the starting node to all other nodes.

**5. Graph Visualization**  
- The Karate Club graph, **color-coded based on the calculated distances**, is drawn. The `nx.draw_spring` function is used to visualize the graph with a spring layout.  
- The color of each node is determined by its distance from the starting node.

**Console Output**  
**List displayed at the top**  
The distances from the starting node to each node are shown. For example, if `print(dist)` outputs `[0, 1, 1, 2, ...]`, it indicates that the distance from node 0 is 0, the distances from nodes 1 and 2 are 1, and the distance from node 3 is 2.

**Execution of BFS and Display of Results**  
The edges explored by BFS starting from node 0 are listed.

**Display of BFS Results with Different Depth Limits**  
A list of newly explored edges is shown for each depth limit: depth 1, depth 2, depth 3, depth 4, and depth 5.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the Karate Club graph
# Generate Zachary's Karate Club graph.
G = nx.karate_club_graph()

# Initialization
# known: Records whether each node has been visited.
# dist: Records the distance of each node from the starting node.
# colors: Defines the color of nodes based on their distance.
# color_map: Stores the color for each node (initially all black).
known = [0] * nx.number_of_nodes(G)
dist = [-1] * nx.number_of_nodes(G)
colors = ['red', 'gray', 'lightgray', 'white']
color_map = ['black'] * nx.number_of_nodes(G)

# Set the starting node
# Set the distance of the starting node to 0 and assign its color.
# Also, mark the starting node as visited.
start = 0
dist[start] = 0
color_map[start] = colors[dist[start]]
known[start] = 1

# Execute Breadth-First Search (BFS)
# Perform node visitation and update distances.
d = 0
while sum(known) != nx.number_of_nodes(G):
    for n in nx.nodes(G):
        # Explore nodes with the current distance of d.
        if nx.shortest_path_length(G, start, n) == d:
            for nb in G.neighbors(n):
                # If the neighboring node is unvisited, update the distance and set the color.
                if known[nb] != 1:
                    dist[nb] = d + 1
                    color_map[nb] = colors[dist[nb]]
                    known[nb] = 1
    d = d + 1

# Display the distance of each node
print(dist)

# Graph visualization
# Set node colors and draw the Karate Club graph.
plt.figure(figsize=(5, 5))
nx.draw_spring(G, node_size=400, node_color=color_map, with_labels=True, font_weight='bold')
plt.show()

# Breadth-First Search (BFS)
print("BFS:", list(nx.bfs_edges(G, source=0)))

# BFS results with different depth limits
d1 = list(nx.bfs_edges(G, source=0, depth_limit=1))
print("depth 1:", d1)
d2 = list(nx.bfs_edges(G, source=0, depth_limit=2))
print("depth 2:", list(set(d2) - set(d1)))
d3 = list(nx.bfs_edges(G, source=0, depth_limit=3))
print("depth 3:", list(set(d3) - set(d2)))
d4 = list(nx.bfs_edges(G, source=0, depth_limit=4))
print("depth 4:", list(set(d4) - set(d3)))
d5 = list(nx.bfs_edges(G, source=0, depth_limit=5))
print("depth 5:", list(set(d5) - set(d4)))
```

## Node Reachability Analysis Using Depth-Limited Breadth-First Search (BFS)

First, please download the "road-chesapeake.mtx" file from the following URL:  
https://networkrepository.com/road-chesapeake.php

This script uses the Breadth-First Search (BFS) algorithm to visualize node reachability within a graph under different depth limits. The graph data is first read from the specified .mtx file and then converted into a graph object using the NetworkX library.

Next, BFS is executed with four different depth limits (1, 2, 3, and 4). For each limit, the BFS result is used to visualize the edges reached during the search. The red edges in the plots represent the connections between nodes that were reached by the BFS for each depth limit.

By visually analyzing the BFS results for different depth limits, you can intuitively understand how the search range changes. This provides insight into the characteristics of breadth-first search and how limiting the depth affects the exploration results.

Finally, the script examines the connected components of the entire graph, displaying the total number of components and the size of the largest connected component. This deepens the understanding of the graph structure and offers a more comprehensive evaluation of the BFS results.

By running this script, you can visually compare and analyze the impact of depth-limited BFS on node reachability, gaining important insights into graph traversal.

```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.io import mmread

# Load the .mtx file (please specify the path according to your environment)
matrix = mmread(r"D:\Python\NetworkX\road-chesapeake\road-chesapeake.mtx")

# Convert to NetworkX graph object
G = nx.from_scipy_sparse_array(matrix)

# Prepare for graph visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
axes = axes.flatten()
pos = nx.spring_layout(G, seed=42)  # Force-directed layout

# Display BFS results for each depth limit
depth_limits = [1, 2, 3, 4]  # Depth limits for each subplot
reached_nodes_count = []

for i, depth_limit in enumerate(depth_limits):
    bfs_edges = list(nx.bfs_edges(G, source=0, depth_limit=depth_limit))
    bfs_tree = nx.DiGraph(bfs_edges)
    reached_nodes = set(n for edge in bfs_edges for n in edge)
    reached_nodes_count.append(len(reached_nodes))
    
    print(f"Depth Limit: {depth_limit}")
    print(f"BFS Edges: {bfs_edges}")
    print(f"Reached Nodes: {reached_nodes}\n")
    
    # Draw BFS results on the subplot
    ax = axes[i]
    ax.set_title(f"Depth Limit: {depth_limit}")
    
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color='blue', alpha=0.7, ax=ax)
    nx.draw_networkx_edges(G, pos, alpha=0.1, ax=ax)
    nx.draw_networkx_edges(bfs_tree, pos, edge_color='red', width=2, alpha=0.7, ax=ax)

plt.suptitle("BFS with Different Depth Limits")
plt.show()

# Plot the number of reached nodes
plt.figure(figsize=(10, 6))
plt.plot(depth_limits, reached_nodes_count, marker='o')
plt.title("Number of Nodes Reached by BFS with Different Depth Limits")
plt.xlabel("Depth Limit")
plt.ylabel("Number of Nodes Reached")
plt.xticks(depth_limits)
plt.grid(True)
plt.show()

# Examine connected components
connected_components = list(nx.connected_components(G))
num_connected_components = len(connected_components)
largest_component = max(connected_components, key=len)

print(f"Number of connected components: {num_connected_components}")
print(f"Size of largest component: {len(largest_component)}")
```

## Depth-First Search (DFS)

**Depth-First Search (DFS)** is a graph and tree traversal algorithm characterized by its approach of exploring "deeply."  
In DFS, the search starts from a given starting point (the source) and proceeds as far as possible along one path before backtracking. The algorithm explores a single path to its end, and upon reaching a dead-end, it retraces its steps to explore other available paths.

As a concrete example, imagine you're standing at the entrance of a maze. You pick one path and follow it as far as it will go. When you hit a dead-end, you backtrack and look for another path to continue. This process repeats until you've explored the entire maze.

Another relatable example is navigating a file system on a computer. Picture a folder containing subfolders, which in turn may contain more subfolders, forming a hierarchical structure. DFS involves opening one folder, going as deep as possible into its subfolders, and when no further subfolders are found, going back to the higher-level folder to explore the remaining ones.

The steps of DFS can be summarized as follows:

1. **Add the starting node to a stack.** A stack is a data structure where the last element added is the first to be removed (LIFO: Last In, First Out).
2. **While the stack is not empty**, repeat the following steps:
   - Remove the top node from the stack and mark it as "visited."
   - Add all unvisited adjacent nodes of the current node to the stack.

By following these steps, DFS explores one path deeply before backtracking to explore alternative paths.

An example application of DFS is solving puzzles. For instance, finding the exit in a maze or considering potential future moves in a game like chess are examples of DFS in action. In computer science, DFS is commonly used to find connected components in a graph or for web crawling to explore links between web pages. Since DFS thoroughly explores a path, it's particularly effective when you want to deeply investigate specific routes.

Thus, **Depth-First Search (DFS)** is an intuitive and versatile algorithm that plays a crucial role in various fields.

This algorithm follows the strategy of "going as far as possible." In other words, it keeps moving deeply into the graph, visiting unvisited nodes, until no more paths can be explored. That's why it is called **"Depth-First Search"**.

---

By running the following script, the Karate Club graph will be visualized, and the DFS traversal results will be output to the console.  
Specifically, the two functions **nx.dfs_edges(G, source=0)** and **nx.dfs_preorder_nodes(G, source=0)** are used to perform depth-first search on graph G starting from node 0.

**1. Generation of the Karate Club Graph**  
- The Karate Club graph is generated using NetworkX's **karate_club_graph**.

**2. Execution of DFS**  
- The **nx.dfs_edges()** function is used to obtain a list of edges from the depth-first search starting at node 0, which is then displayed in the console.  
- The **nx.dfs_preorder_nodes()** function is used to obtain the order in which the nodes are visited in the depth-first search starting at node 0, and this order is displayed in the console.

**3. Visualization of the Graph:**  
- The **nx.draw_spring()** function is used to draw the Karate Club graph using a spring layout. The nodes are displayed in red, and labels are also added to the nodes.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Generate the Karate Club graph
G = nx.karate_club_graph()

# Execute Depth-First Search (DFS) and display the results
dfs_edges = list(nx.dfs_edges(G, source=0))
dfs_nodes = list(nx.dfs_preorder_nodes(G, source=0))

print("DFS Edges:", dfs_edges)
print("Traversed Nodes:", dfs_nodes)

# Graph drawing to highlight DFS path
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 10))
nx.draw(G, pos, node_size=400, node_color='lightblue', with_labels=True, font_weight='bold')

# Highlight the DFS path in red
edge_colors = ['red' if edge in dfs_edges or (edge[1], edge[0]) in dfs_edges else 'black' for edge in G.edges()]
nx.draw_networkx_edges(G, pos, edge_color=edge_colors)
plt.title("Depth-First Search (DFS) on Karate Club Graph")
plt.show()
```

## Analysis of Node Search Behavior Using Depth-Limited DFS

This script is designed to visualize and analyze node reachability within a graph using the Depth-First Search (DFS) algorithm with varying depth limits. First, the data is loaded from a specified `.mtx` file and converted into a graph object using the NetworkX library.

Next, four different depth limits (1, 2, 3, 4) are set, and DFS is executed for each limit to observe how nodes are reached. The results of the DFS are plotted in subplots, with the reached nodes highlighted by red edges, allowing for a visual understanding of which nodes were reached.

The number of nodes reached at each depth limit is plotted, showing how many nodes can be reached as the depth limit increases. This helps to understand the behavior of DFS and the effect of depth limits on the exploration process.

Finally, the script examines the connected components of the entire graph, displaying the total number of components and the size of the largest connected component. This provides a broader understanding of the graph’s structure and the overall results of the DFS exploration.

By running this script, you can visually compare and analyze the impact of depth limits on the DFS algorithm’s search range, gaining important insights into graph traversal concepts.

```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.io import mmread

def dfs_with_depth_limit(G, start, depth_limit):
    visited = set()
    stack = [(start, 0)]
    edges = []
    
    while stack:
        node, depth = stack.pop()
        if node not in visited and depth <= depth_limit:
            visited.add(node)
            if depth < depth_limit:
                for neighbor in G[node]:
                    if neighbor not in visited:
                        stack.append((neighbor, depth + 1))
                        edges.append((node, neighbor))
    return edges

# Load the .mtx file
matrix = mmread(r"D:\Python\NetworkX\road-chesapeake\road-chesapeake.mtx")

# Convert to a NetworkX graph object
G = nx.from_scipy_sparse_array(matrix)

# Prepare for graph visualization
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
axes = axes.flatten()
pos = nx.spring_layout(G, seed=42)  # Force-directed layout

# Display DFS results for each depth limit
depth_limits = [1, 2, 3, 4]  # Depth limits for each subplot
reached_nodes_count = []

for i, depth_limit in enumerate(depth_limits):
    dfs_edges = dfs_with_depth_limit(G, 0, depth_limit)
    dfs_tree = nx.DiGraph(dfs_edges)
    reached_nodes = set(n for edge in dfs_edges for n in edge)
    reached_nodes_count.append(len(reached_nodes))
    
    print(f"Depth Limit: {depth_limit}")
    print(f"DFS Edges: {dfs_edges}")
    print(f"Reached Nodes: {reached_nodes}\n")
    
    # Draw DFS results on the subplot
    ax = axes[i]
    ax.set_title(f"Depth Limit: {depth_limit}")
    
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color='blue', alpha=0.7, ax=ax)
    nx.draw_networkx_edges(G, pos, alpha=0.1, ax=ax)
    nx.draw_networkx_edges(dfs_tree, pos, edge_color='red', width=2, alpha=0.7, ax=ax)

plt.suptitle("DFS with Different Depth Limits")
plt.show()

# Plot the number of nodes reached
plt.figure(figsize=(10, 6))
plt.plot(depth_limits, reached_nodes_count, marker='o')
plt.title("Number of Nodes Reached by DFS with Different Depth Limits")
plt.xlabel("Depth Limit")
plt.ylabel("Number of Nodes Reached")
plt.xticks(depth_limits)
plt.grid(True)
plt.show()

# Examine connected components
connected_components = list(nx.connected_components(G))
num_connected_components = len(connected_components)
largest_component = max(connected_components, key=len)

print(f"Number of connected components: {num_connected_components}")
print(f"Size of largest component: {len(largest_component)}")
```

## Comparison of Graph Search Algorithms (BFS vs DFS)

This script compares how nodes are reached within a graph using the Breadth-First Search (BFS) and Depth-First Search (DFS) algorithms. First, data is loaded from a specified .mtx file and converted into a graph object using the NetworkX library.

Next, four depth limits are set, and the results of BFS and DFS are obtained for each limit.
- For BFS, edges from the starting node are retrieved for each depth limit and displayed as a directed graph.
- For DFS, edges are explored while considering the depth limit, and similarly, the results are displayed as a directed graph.

The results of BFS and DFS for each depth limit are drawn in subplots, visually comparing how many nodes were reached. This allows us to observe the differences in the behavior of BFS and DFS under different depth limits.

Additionally, the number of nodes reached at each depth limit is plotted, and the number of reached nodes in BFS and DFS is compared.

Finally, the connected components of the entire graph are examined, showing the number and size of the largest connected component. This helps to deepen the understanding of the graph's structure and the characteristics of the search algorithms.


```python
import networkx as nx
import matplotlib.pyplot as plt
from scipy.io import mmread

def dfs_with_depth_limit(G, start, depth_limit):
    visited = set()
    stack = [(start, 0)]
    edges = []
    
    while stack:
        node, depth = stack.pop()
        if node not in visited and depth <= depth_limit:
            visited.add(node)
            if depth < depth_limit:
                for neighbor in G[node]:
                    if neighbor not in visited:
                        stack.append((neighbor, depth + 1))
                        edges.append((node, neighbor))
    return edges

# Load the .mtx file
matrix = mmread(r"D:\Python\NetworkX\road-chesapeake\road-chesapeake.mtx")

# Convert to a NetworkX graph object
G = nx.from_scipy_sparse_array(matrix)

# Prepare for graph display
fig, axes = plt.subplots(4, 2, figsize=(15, 30))
axes = axes.flatten()
pos = nx.spring_layout(G, seed=42)  # Force-directed layout

# Display BFS and DFS results for each depth limit
depth_limits = [1, 2, 3, 4]  # Depth limits for each subplot
bfs_reached_nodes_count = []
dfs_reached_nodes_count = []

for i, depth_limit in enumerate(depth_limits):
    # BFS
    bfs_edges = list(nx.bfs_edges(G, source=0, depth_limit=depth_limit))
    bfs_tree = nx.DiGraph(bfs_edges)
    bfs_reached_nodes = set(n for edge in bfs_edges for n in edge)
    bfs_reached_nodes_count.append(len(bfs_reached_nodes))
    
    print(f"BFS - Depth Limit: {depth_limit}")
    print(f"BFS Edges: {bfs_edges}")
    print(f"BFS Reached Nodes: {bfs_reached_nodes}\n")
    
    # Plot BFS results in the subplot
    ax = axes[2*i]
    ax.set_title(f"BFS - Depth Limit: {depth_limit}")
    
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color='blue', alpha=0.7, ax=ax)
    nx.draw_networkx_edges(G, pos, alpha=0.1, ax=ax)
    nx.draw_networkx_edges(bfs_tree, pos, edge_color='red', width=2, alpha=0.7, ax=ax)
    
    # DFS
    dfs_edges = dfs_with_depth_limit(G, 0, depth_limit)
    dfs_tree = nx.DiGraph(dfs_edges)
    dfs_reached_nodes = set(n for edge in dfs_edges for n in edge)
    dfs_reached_nodes_count.append(len(dfs_reached_nodes))
    
    print(f"DFS - Depth Limit: {depth_limit}")
    print(f"DFS Edges: {dfs_edges}")
    print(f"DFS Reached Nodes: {dfs_reached_nodes}\n")
    
    # Plot DFS results in the subplot
    ax = axes[2*i + 1]
    ax.set_title(f"DFS - Depth Limit: {depth_limit}")
    
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color='blue', alpha=0.7, ax=ax)
    nx.draw_networkx_edges(G, pos, alpha=0.1, ax=ax)
    nx.draw_networkx_edges(dfs_tree, pos, edge_color='red', width=2, alpha=0.7, ax=ax)

plt.suptitle("BFS and DFS with Different Depth Limits")
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Plot the number of reached nodes
plt.figure(figsize=(10, 6))
plt.plot(depth_limits, bfs_reached_nodes_count, marker='o', label='BFS')
plt.plot(depth_limits, dfs_reached_nodes_count, marker='x', label='DFS')
plt.title("Number of Nodes Reached by BFS and DFS with Different Depth Limits")
plt.xlabel("Depth Limit")
plt.ylabel("Number of Nodes Reached")
plt.xticks(depth_limits)
plt.legend()
plt.grid(True)
plt.show()

# Investigate connected components
connected_components = list(nx.connected_components(G))
num_connected_components = len(connected_components)
largest_component = max(connected_components, key=len)

print(f"Number of connected components: {num_connected_components}")
print(f"Size of largest component: {len(largest_component)}")
```

## Dijkstra's Algorithm

**Dijkstra's Algorithm** is a method for finding the shortest paths from a specific point in a graph to all other points. The basic idea is to **start from the initial point and gradually approach the destination by selecting the path with the lowest cost**.

For example, consider giving directions on a map. When traveling from one location to another, you first check the distance of all the paths directly accessible from the starting point and choose the shortest one to proceed to the next location. From there, you select the next shortest path, and by repeating this process, you can eventually find the shortest path from the starting point to the destination.

The same concept applies to optimizing delivery routes. When a delivery person has multiple destinations, they find the most efficient order by first choosing the closest destination, then selecting the next closest location from there, ultimately determining the shortest route for all deliveries.

The specific steps of Dijkstra's Algorithm are as follows:

1. Set the distance from the starting point to 0 and the distance to all other points to infinity.
2. Select the starting point from the set of unprocessed points.
3. **For all points adjacent to the current point, calculate the path length.** If the new path is shorter, update the distance to that point.
4. After updating the distances to adjacent points, mark the current point as processed and remove it from the set of unprocessed points.
5. As long as there are unprocessed points remaining, select the point with the shortest distance and repeat steps 3 and 4.

However, if **all edge weights in the graph are the same positive value, using breadth-first search (BFS) is more efficient than Dijkstra's Algorithm**. In such cases, the shortest path can be computed in linear time.

Visualizing Dijkstra's Algorithm offers several advantages:

**Intuitive understanding of the shortest path**: It provides a clear view of how the shortest path from a specific starting point to all other nodes progresses, making the construction process of the shortest path easy to understand.

**Tracking the algorithm's behavior**: You can visually follow the process of updating the shortest distances step by step.

**Understanding of costs**: It visually demonstrates how the cumulative costs (distances) to each node change, making it easier to understand the process of minimizing costs when determining the shortest path.

**Comprehension of the weighted structure of the graph**: It helps in understanding how edge weights influence the determination of the shortest path and provides a deeper grasp of the graph’s weighted structure.

Thus, while Dijkstra's Algorithm has broad applications, there may be cases where other algorithms are more suitable depending on the specific conditions of the problem. The choice should be considered based on the nature of the problem.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import functools
import operator

# Create the graph
G = nx.Graph()
G.add_nodes_from(range(0, 6))  # Add nodes from 0 to 5
G.add_weighted_edges_from([(0, 1, 7), (0, 2, 9), (0, 5, 14), (1, 2, 10), (1, 3, 15), (2, 3, 11), (2, 5, 2), (3, 4, 6), (4, 5, 9)])

# Draw the graph
plt.figure(figsize=(5, 5))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d["weight"] for u, v, d in G.edges(data=True)})
plt.axis('off')
plt.show()

# Initialize Dijkstra's Algorithm
start_node = 1
dist_estimate = [float('inf')] * nx.number_of_nodes(G)
dist_certainty = [0] * nx.number_of_nodes(G)
dist_estimate[start_node] = 0

# Execute Dijkstra's Algorithm
while any(dc == 0 for dc in dist_certainty):
    print(f"Estimated distances: {dist_estimate}")
    print(f"Confirmed distances: {dist_certainty}")
    min_v = float('inf')
    for n in range(len(dist_estimate)):
        if not dist_certainty[n] and dist_estimate[n] <= min_v:
            min_v = dist_estimate[n]
            min_id = n
    dist_certainty[min_id] = 1
    for nb in G.neighbors(min_id):
        new_estimate = G[min_id][nb]['weight'] + dist_estimate[min_id]
        if new_estimate < dist_estimate[nb]:
            dist_estimate[nb] = new_estimate

# Output the final estimated and confirmed distances
print(f"Final estimated distances: {dist_estimate}")
print(f"Confirmed distances: {dist_certainty}")
```

## Bellman-Ford Algorithm

The **Bellman-Ford Algorithm** is an algorithm for finding the shortest path from a given point to all other points in a graph. **The key feature of this algorithm is its ability to handle graphs with edges that have negative weights**.

The core of the Bellman-Ford algorithm involves repeatedly checking each edge to see if taking that edge shortens the path. By repeating this process for all edges in the graph, the algorithm eventually finds the shortest path.

For example, imagine finding the shortest route from the school gate to a classroom. First, you record the distances of all direct routes from the gate, and then you check each recorded route to see if the path can be shortened. By repeating this process for all routes in the school, you can find the shortest path to the classroom.

Another everyday example is grocery shopping. Suppose you want to find the shortest path to several items on your shopping list in a supermarket. First, you calculate the shortest route to each item using a store map, and then you repeatedly check if those routes can be made even shorter. By doing this, you can find the most efficient way to complete your shopping.

The specific steps of the Bellman-Ford algorithm are as follows:

1. **Set the distance from the starting point to 0**, and set the distances to all other points to infinity.
2. **Repeat the following operation for every edge in the graph** (this is called "edge relaxation"):
   - For each edge, if taking that edge shortens the path, update the distance to the destination node.
3. **Repeat this process for all nodes (V - 1 times)** to find the shortest paths.
4. Finally, check all edges one more time to detect negative cycles.

**The Bellman-Ford algorithm is particularly effective for finding shortest paths in graphs with negative weights.**  
The time complexity of the Bellman-Ford algorithm is $O(VE)$, where $V$ is the number of nodes and $E$ is the number of edges. This is because every edge is checked (V - 1) times.

**Comparison with Dijkstra's Algorithm**  
Dijkstra's algorithm cannot handle negative weights but is faster (with a time complexity of $O((V+E)\log V)$ when using a priority queue). On the other hand, the Bellman-Ford algorithm is more versatile because it can handle negative weights, but its time complexity is higher.

**Detection of Negative Cycles**  
The Bellman-Ford algorithm is the only shortest path algorithm that can detect negative cycles (cycles where the total weight is negative when traversing the cycle). This is achieved by performing one additional iteration after V - 1 iterations and checking if any distances are still being updated.

---

When you run this script, the shortest paths from the starting point to other nodes in the specified graph will be visualized.

1. **Graph Creation**  
   A directed graph is created using `networkx.DiGraph()`, and edges along with their weights are added.

2. **Implementation of the Bellman-Ford Algorithm**  
   The `bellman_ford` function is defined, performing initialization, edge relaxation (checking and updating if the path can be shortened through each edge), and detection of negative cycles.

3. **Shortest Path Calculation**  
   The Bellman-Ford algorithm is applied with the starting point set to 'A', and the shortest distances and predecessors for each node are calculated.

4. **Displaying Results**  
   The calculated results are shown in the console, and the graph is drawn using Matplotlib.

5. **Visualization of Shortest Paths**  
   The shortest paths are highlighted in red, allowing for a visual confirmation of the shortest paths from the starting point to each node.

```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import rcParams

# Set font to English font
rcParams['font.family'] = 'DejaVu Sans'

def bellman_ford(graph, source):
    # Initialize distance and predecessor for each node
    distance = {node: float('inf') for node in graph.nodes()}
    predecessor = {node: None for node in graph.nodes()}
    distance[source] = 0

    # Bellman-Ford relaxation process
    for _ in range(len(graph.nodes()) - 1):
        for u, v, data in graph.edges(data=True):
            weight = data['weight']
            if distance[u] + weight < distance[v]:
                distance[v] = distance[u] + weight
                predecessor[v] = u

    # Detect negative-weight cycle
    for u, v, data in graph.edges(data=True):
        weight = data['weight']
        if distance[u] + weight < distance[v]:
            raise ValueError("Graph contains a negative-weight cycle")

    return distance, predecessor

# Create the graph
G = nx.DiGraph()
edges = [
    ('A', 'B', 1),
    ('A', 'C', 4),
    ('B', 'C', -3),
    ('B', 'D', 2),
    ('C', 'E', 2),
    ('D', 'E', 3)
]
G.add_weighted_edges_from(edges)

# Calculate shortest paths from the source node
source = 'A'
distance, predecessor = bellman_ford(G, source)

# Output to the console
print("Distance from source '{}' to each node:".format(source))
for node, dist in distance.items():
    print("  Node {}: Distance {}".format(node, dist))

print("\nPredecessors of each node (the previous node in the shortest path):")
for node, pred in predecessor.items():
    print("  Node {}: Predecessor {}".format(node, pred))

# Manually set the position of the nodes
pos = {
    'A': (0, 0),
    'B': (2, -1),
    'C': (1, 1),
    'D': (3, 1),
    'E': (4, 2)
}

# Draw the graph
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=1000, font_size=16, font_weight='bold', arrowsize=20)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=12, label_pos=0.3)

# Draw the shortest paths
for node in G.nodes():
    if node != source:
        path = []
        current = node
        while current is not None:
            path.append(current)
            current = predecessor[current]
        path = path[::-1]
        edges_in_path = [(path[i], path[i + 1]) for i in range(len(path) - 1)]
        nx.draw_networkx_edges(G, pos, edgelist=edges_in_path, edge_color='red', width=3, arrowsize=25)

plt.title("Bellman-Ford Algorithm: Shortest Paths from Source '{}'".format(source), fontsize=20)
plt.show()
```

## Max-Flow Min-Cut Theorem

The **Max-Flow Min-Cut Theorem** is one of the most important theorems in graph theory and provides a method to optimize the flow within a network. This theorem is used to find the maximum flow from a source to a sink in a network, and it states that the maximum flow is equal to the minimum cut (the total capacity of edges that divide the network into two parts).

The basic idea of this theorem is to **find the maximum possible flow that can pass through the network**. This maximum flow matches the capacity of the minimum cut that separates the source and the sink.

For a concrete example, consider the flow of a river. Suppose there is a dam along the river. The maximum flow of the river is determined by how much water the dam can allow to pass (the capacity of the dam). The amount of water the dam can pass becomes the maximum flow of the river, and any excess water will be blocked by the dam. In this case, the river’s flow (maximum flow) is restricted by the dam (the minimum cut).

Another relatable example is a pipeline system. Imagine multiple pipelines running from a factory to consumers, each with a maximum capacity. The maximum supply from the factory to the consumers is determined by the narrowest pipe in the system. This narrow pipe acts as the minimum cut that restricts the maximum flow.

The specific steps to calculate the Max-Flow Min-Cut Theorem are as follows:

1. To solve the maximum flow problem, first examine all possible paths from the source to the sink.
2. Assign flow along these paths, ensuring that no edge exceeds its capacity.
3. Adjust the flow to maximize the total flow from the source to the sink. At this point, the maximum flow will equal the capacity of the minimum cut that separates the source and sink.

Applications of the Max-Flow Min-Cut Theorem include traffic system optimization. For example, it can be used to calculate the maximum number of cars that can pass through a specific route in a city’s traffic network, which can help develop measures to reduce congestion. It is also used in communication networks to design systems that allow efficient data flow. In the logistics industry, the theorem is used to optimize the transportation routes for goods, playing an important role in efficient transportation planning.

The Max-Flow Min-Cut Theorem is particularly effective for optimizing the flow within a network.

### Visualization of the Max-Flow Min-Cut Theorem provides the following advantages:
**Intuitive understanding of flow distribution**  
・Visualization allows for a clear view of the flow paths from the source (starting point) to the sink (end point), as well as the amount of flow passing through each edge. This makes it easy to intuitively understand how the overall flow is distributed.

**Identifying bottlenecks**  
・Bottleneck edges or nodes that restrict the flow can be easily identified. This makes it clear which parts of the network need to be strengthened to improve capacity and efficiency.

**Understanding the relationship between flow and cut**

**Performance analysis of the network**  
・It allows for a visual evaluation of the performance and limitations of the network. For example, in communication or transportation networks, it’s easy to analyze how much of the current maximum flow is using the network’s capacity and where there is room for improvement.

**Maximum Flow** aims to find the maximum amount of data or materials that can be transferred from a specific source to a sink. This is particularly important when each edge in the network has a specific capacity (or weight).

On the other hand, the **Minimum Cut** aims to find the smallest set of edges that need to be cut to divide the network into two parts. The minimum cut is the one whose capacity (the sum of the weights of the edges that form the cut) is the smallest.

The Max-Flow Min-Cut Theorem states that the maximum flow in a network is always equal to the capacity of the minimum cut in that network. This is because the flow is always constrained by the edge with the smallest capacity along the path, and this smallest edge becomes the minimum cut.

Thus, the relationship between the maximum flow and the minimum cut is very useful for understanding the nature of the network and its constraints (bottlenecks).

```python
import networkx as nx
import matplotlib.pyplot as plt

# Explanation of the Max-Flow Min-Cut Theorem
"""
Max-Flow Min-Cut Theorem:
In a network, the value of the maximum flow from the source to the sink is equal to the capacity of the minimum cut that divides the source and the sink.
Here,
- **Maximum Flow**: The maximum amount of flow that can be sent from the source to the sink.
- **Minimum Cut**: The set of edges that divide the source and the sink, with the smallest total capacity.
"""

# Definition of the graph
G = nx.DiGraph()
G.add_edge('s', 'a', capacity=3.0)
G.add_edge('s', 'b', capacity=1.0)
G.add_edge('a', 'c', capacity=3.0)
G.add_edge('b', 'c', capacity=5.0)
G.add_edge('b', 'd', capacity=4.0)
G.add_edge('d', 'e', capacity=2.0)
G.add_edge('c', 't', capacity=2.0)
G.add_edge('e', 't', capacity=3.0)

# Calculation of the maximum flow
flow_value, flow_dict = nx.maximum_flow(G, 's', 't')

# Calculation of the minimum cut
cut_value, partition = nx.minimum_cut(G, 's', 't')
reachable, non_reachable = partition

# Calculation of the cut set
cutset = set()
for u in reachable:
    for v in G[u]:
        if v in non_reachable:
            cutset.add((u, v))

# Graph visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

# Subplot 1: Visualization of the maximum flow
pos = nx.spring_layout(G, seed=42)
edge_colors = []
edge_widths = []
for u, v in G.edges():
    flow = flow_dict[u][v]
    capacity = G[u][v]['capacity']
    edge_colors.append('green' if flow > 0 else 'black')
    edge_widths.append(flow / capacity * 5)

nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=700, ax=axes[0])
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=edge_widths, arrows=True, arrowstyle='->', arrowsize=20, ax=axes[0])
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{G[u][v]['capacity']}" for u, v in G.edges()}, font_color='red', font_family='sans-serif', ax=axes[0])
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold', font_family='sans-serif', ax=axes[0])
axes[0].set_title("Maximum Flow Visualization", fontsize=16, fontdict={'family': 'sans-serif'})
axes[0].axis('off')

# Subplot 2: Visualization of the minimum cut
node_colors = []
for node in G.nodes():
    if node in reachable:
        node_colors.append('lightgreen')
    else:
        node_colors.append('lightcoral')

edge_colors = []
for u, v in G.edges():
    if (u, v) in cutset:
        edge_colors.append('red')
    else:
        edge_colors.append('black')

nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=700, ax=axes[1])
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2, arrows=True, arrowstyle='->', arrowsize=20, ax=axes[1])
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"{G[u][v]['capacity']}" for u, v in G.edges()}, font_color='red', font_family='sans-serif', ax=axes[1])
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold', font_family='sans-serif', ax=axes[1])
axes[1].set_title("Minimum Cut Visualization", fontsize=16, fontdict={'family': 'sans-serif'})
axes[1].axis('off')

plt.tight_layout()
plt.show()

# Organizing console output
print("=== Maximum Flow Calculation ===")
print(f"Maximum flow from source 's' to sink 't': {flow_value}")
print("Flow through each edge:")
for u in flow_dict:
    for v in flow_dict[u]:
        if flow_dict[u][v] > 0:
            print(f"  {u} -> {v}: {flow_dict[u][v]}")
print()

print("=== Minimum Cut Calculation ===")
print(f"Minimum cut capacity: {cut_value}")
print(f"Nodes reachable from source 's': {reachable}")
print(f"Nodes not reachable: {non_reachable}")
print("Edges included in the cut set:")
for edge in sorted(cutset):
    capacity = G[edge[0]][edge[1]]['capacity']
    print(f"  {edge[0]} -> {edge[1]}: Capacity = {capacity}")

# Total capacity of the cut set
cutset_capacity = sum(G[u][v]['capacity'] for (u, v) in cutset)
print(f"\nTotal capacity of the cut set: {cutset_capacity}")
print(f"Matches minimum cut capacity: {cut_value == cutset_capacity}")
print()

# Displaying the conclusion
print("=== Conclusion ===")
print("The value of the maximum flow matches the capacity of the minimum cut. This demonstrates the Max-Flow Min-Cut Theorem.")
```

## A* Algorithm

The A* algorithm is an efficient approach to solving the shortest path problem by combining the current cost (distance) and the estimated remaining cost (heuristic). The strength of this algorithm lies in its combination of the certainty of Dijkstra's algorithm with the efficiency of the heuristic. The heuristic function provides an estimated cost to the goal, guiding the search in an efficient direction.

For example, imagine you are walking through a city to visit a friend's house. You want to arrive as quickly as possible, so you choose the closest path from your current location while also estimating the distance to your friend’s house. In doing so, you're using the A* algorithm. You first select the closest path from the starting point and then decide which path to take next by considering the estimated remaining distance to the goal.

The A* algorithm has a wide range of applications, from simple grid-based pathfinding to more complex graph-based pathfinding.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Generate the graph
G = nx.grid_2d_graph(5, 5)

# Set the node positions
pos = {(x, y): (x, y) for x, y in G.nodes()}

# Set the weights (used as distances)
for (u, v) in G.edges():
    G.edges[u, v]['weight'] = 1

# Calculate the shortest path using the A* algorithm
start = (0, 0)
goal = (4, 4)
path = nx.astar_path(G, start, goal, heuristic=lambda u, v: abs(u[0] - v[0]) + abs(u[1] - v[1]))

# Display the result
print("A* Algorithm Path:", path)

# Draw the graph
plt.figure(figsize=(8, 8))

# Draw all nodes and edges
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, edge_color='gray')

# Highlight the shortest path
path_edges = list(zip(path, path[1:]))
nx.draw_networkx_nodes(G, pos, nodelist=path, node_color='red', node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2)

plt.title("A* Algorithm - Shortest Path")
plt.show()
```

In theory, if the heuristic function in the A* algorithm is admissible and consistent, both algorithms should yield the same shortest path. However, due to differences in the calculation process and search order, there is a rare chance that different paths may be selected.

Through this script, you can compare the behavior and performance of these two important algorithms in graph theory, and it is particularly useful for understanding the impact of the heuristic function in the A* algorithm.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Set the number of nodes and edge creation probability
num_nodes = 10
probability = 0.3

# Generate an Erdos-Renyi random graph
G = nx.erdos_renyi_graph(num_nodes, probability)

# Set the start and goal nodes
start_node = 0
goal_node = num_nodes - 1

# Generate node positions
pos = nx.spring_layout(G)

# Apply Dijkstra's algorithm to calculate the shortest path
shortest_path = nx.dijkstra_path(G, start_node, goal_node)

# Define the heuristic function (Euclidean distance between nodes)
def heuristic(u, v):
    return ((pos[u][0] - pos[v][0])**2 + (pos[u][1] - pos[v][1])**2) ** 0.5

# Apply A* algorithm to calculate the shortest path
astar_path = nx.astar_path(G, start_node, goal_node, heuristic=heuristic)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Plot the Erdos-Renyi random graph
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=15, ax=axs[0])
axs[0].set_title("Dijkstra Path")

# Highlight and plot the Dijkstra path
path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=2, ax=axs[0])

# Highlight and plot the A* algorithm path
path_edges = list(zip(astar_path, astar_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='blue', width=2, ax=axs[1])
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=15, ax=axs[1])
axs[1].set_title(f"Shortest Path from Node {start_node} to Node {goal_node} using A* Algorithm")

# Add a legend
legend_labels = {
    'Dijkstra Path': 'red',
    'A* Path': 'blue'
}
for label, color in legend_labels.items():
    axs[1].plot([0], [0], color=color, label=label)
axs[1].legend(loc='upper left')

plt.show()
```
