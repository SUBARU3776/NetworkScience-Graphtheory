# Network Models  
To deepen the understanding of networks, it is essential to first grasp the basic mechanisms of how networks are constructed and subsequently analyzed. Networks consist of nodes (elements) and edges (interactions between them). Examples include friendships on social media or links between web pages on the internet.  

In network analysis, various methods are employed to reveal the structure and characteristics of a network. Models such as random graphs, configuration models, scale-free graphs, and small-world graphs each possess unique traits and offer insights into real-world networks.  

These models emphasize specific aspects of networks and provide a theoretical framework. For instance, **random graphs exhibit uniform connectivity**, while **scale-free graphs are characterized by a dynamic structure with hub nodes**. By understanding these models, one can gain a deeper insight into network formation, functionality, and information propagation patterns.

Each graph model exhibits characteristic differences in terms of degree distribution, clustering coefficient, average path length, and rewiring probability. Below are the features of each:

**1. Random Graph (Erdős-Rényi Model)**  
**・** Degree Distribution: Follows a Poisson distribution. All nodes have an equal probability of having edges, leading to a roughly uniform degree among nodes.  
**・** Clustering Coefficient: Low. Since edges are added randomly, triangles (closed triples) are less likely to form among neighboring nodes.  
**・** Average Path Length: Relatively short. The distance between nodes tends to be short, scaling logarithmically with the number of nodes.  
**・** Rewiring Probability: Not specifically defined, but as edges are distributed evenly, the rewiring probability is generally uniform.

**2. Configuration Model**  
**・** Degree Distribution: Can be set to any desired distribution.  
**・** Clustering Coefficient: Relatively low. Connections between nodes depend on the degree distribution, but clustering is generally low.  
**・** Average Path Length: Typically short. The distance between nodes often shortens logarithmically with the number of nodes.  
**・** Rewiring Probability: Not specifically defined, but connections between nodes depend on the degree distribution.

**3. BA Model (Barabási-Albert Model)**  
**・** Degree Distribution: Scale-free network. The degree distribution follows a power law, where a few nodes have very high degrees while most others have low degrees.  
**・** Clustering Coefficient: Moderate. Since new nodes preferentially attach to existing ones, local clustering is observed to some extent.  
**・** Average Path Length: Extremely short. Due to the scale-free nature, the distance between nodes tends to be very small.  
**・** Rewiring Probability: Not specifically defined, but since higher-degree nodes are preferentially connected, edges tend to concentrate around certain nodes.

**4. Small-World (Watts-Strogatz Model)**  
**・** Degree Distribution: Approximates a uniform or Poisson distribution. As the number of nodes increases, the degree distribution often becomes more uniform.  
**・** Clustering Coefficient: High. Many triangles (closed triples) form between neighboring nodes, indicating strong local clustering.  
**・** Average Path Length: Short. The distance between nodes is short, exhibiting the small-world property.  
**・** Rewiring Probability: Varies depending on the degree of randomization. As connections between nodes are partially replaced randomly, the rewiring probability can be adjusted.

**5. Complete Graph**  
**・** Degree Distribution: All nodes have the same degree. If there are $N$ nodes, the degree of each node is $N-1$.  
**・** Clustering Coefficient: 1. Since all nodes are fully connected, triangles are always formed.  
**・** Average Path Length: 1. Since every node is directly connected to every other node, the distance is always 1.  
**・** Rewiring Probability: Not specifically defined, but since every node is connected to every other node directly, the rewiring probability is meaningless.

## Let's Explore the Differences in Degree Distribution  
Degree distribution in a network refers to the distribution of the number of edges (connections) each node has. The degree of a node refers to the number of edges connected to that node. Specifically, degree distribution can be understood as follows:

1. **Definition**  
・ Degree distribution is a statistical representation of how the degrees (number of edges) of nodes are distributed across the network.

2. **Importance**  
・ Degree distribution plays a critical role in understanding the structure of a network. For example, the shape of the degree distribution can help estimate whether a network is random, scale-free, or closer to another specific model.

3. **Examples**  
・ In the case of a scale-free network, the degree distribution typically follows a power law. This means that a small number of hub nodes have very high degrees, while the majority of nodes have low degrees. In contrast, in a random graph, the degree distribution tends to resemble a Poisson distribution.

4. **Analysis Method**  
・ To analyze the degree distribution of a network, a common method is to count the degrees of each node from actual network data and then plot the distribution. The resulting distribution can then be quantified to determine its characteristics.

Understanding degree distribution forms the foundation for grasping how networks are formed and how information propagates through them.

## Random Graph  
**A random graph is a graph created by randomly generating nodes (vertices) and edges.** This method is widely used as a fundamental model to understand the structure and properties of networks.

The creation process is as follows: first, a fixed number of nodes are prepared. Then, for each pair of nodes, the existence of an edge is determined based on a probability. For example, you could flip a coin—if heads, an edge is created, and if tails, no edge is created. By repeating this for all node pairs, a random graph is generated.

The steps for generating a random graph are as follows: First, decide the number of nodes. Then, set the probability of creating an edge between each pair of nodes. Based on this probability, edges are generated randomly, and this operation is repeated for all node pairs. For instance, if the probability of creating an edge is 0.3, there is a 30% chance of creating an edge between each pair of nodes, and a 70% chance of no edge being created. After performing this for all node pairs, the random graph is complete.

Some examples of applications for random graphs are as follows:

**1. Research on Basic Network Properties**: Random graphs are used as a basic model to understand fundamental properties of networks (e.g., average path length, clustering coefficient, degree distribution).  
**2. Simulation of Information Spread**: Random graphs are used to simulate the spread of information or viruses in social networks. This helps predict how information will propagate.  
**3. Modeling Internet Traffic**: Random graphs are also used to model data traffic on the internet, allowing for the evaluation of network load and performance.  
**4. Research on Biological Networks**: Random graphs are employed as basic models for interaction networks of genes or proteins, aiding in the understanding of the underlying structure of biological processes.

Random graphs are a powerful tool for understanding the basic properties of networks and are widely applied in various fields.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Generate an ER graph (Erdős-Rényi graph - Random graph)
n = 100
p = 0.1
er = nx.erdos_renyi_graph(n, p)

# Graph visualization
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
nx.draw(er, node_size=10, node_color='red')
plt.title("Erdős-Rényi Graph")

# Output basic characteristics
print("erdos_renyi_graph")
print("Number of nodes: ", nx.number_of_nodes(er))
print("Number of edges:", nx.number_of_edges(er))
average_degree = nx.number_of_edges(er) * 2 / nx.number_of_nodes(er)
print("Average degree: ", '{:.3f}'.format(average_degree))

# Plot degree distribution
degree_distribution = nx.degree_histogram(er)
plt.subplot(1, 3, 2)
plt.bar(range(len(degree_distribution)), degree_distribution, color='blue')
plt.title("Degree Distribution")

# Comparison with Poisson distribution
lambda_ = n * p  # Theoretical average degree (λ = np)
poisson_dist = poisson.pmf(np.arange(len(degree_distribution)), lambda_) * n

plt.subplot(1, 3, 3)
plt.bar(range(len(degree_distribution)), degree_distribution, color='blue', alpha=0.6, label="Actual")
plt.plot(poisson_dist, 'ro-', label="Poisson (λ=np)")
plt.title("Degree Distribution vs Poisson")
plt.legend()

# Adjust spacing between subplots
plt.subplots_adjust(wspace=0.4)

plt.show()
```

## About the Poisson Distribution  
The Poisson distribution is a tool used to predict the number of "rare events" occurring within a fixed period of time or space. For example, imagine a call center receiving an average of 10 calls per hour, or a flower field where an average of 5 flowers bloom per square meter.

Here, "rare events" refer to occurrences that are possible within a given time or space but have a low probability of happening. Examples include the number of phone calls or the number of flowers blooming, both of which occur randomly.

The probability mass function of the Poisson distribution is expressed as follows:<br>
$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}$

Where:
- $k$ is the number of observed events (e.g., the number of phone calls in a given period or the number of flowers in a given area),
- $e$ is the base of the natural logarithm (a mathematical constant approximately equal to 2.71828),
- $λ$ is the average number of events occurring in a unit time or area,
- $k!$ is the factorial of $k$, representing the product of all integers from $k$ down to 1.

### Characteristics of the Poisson Distribution  
- **Mean and Variance**: In a Poisson distribution, both the mean and variance are equal to $λ$.  
- **Applicability to Infinite Events**: The Poisson distribution is often applied when events occur "infinitely often" but with "low probability" at each instance.  
- **Parameter**: $λ$ indicates the average number of occurrences and determines the shape of the distribution.

Using this formula, one can calculate the probability of an event occurring $k$ times within a specific time or space. In this way, the Poisson distribution serves as a powerful tool for predicting the probability of rare events.

## Comparing Random Graph Generation with the Poisson Distribution  
Comparing random graph generation with the Poisson distribution is a common method for understanding the structure of networks and quantifying their properties. Specifically, it is useful in the following ways:

1. **Distribution**  
・ The **degree distribution** (the distribution of the number of edges per node) of random graphs is theoretically expected to follow a **Poisson distribution**. Therefore, by comparing the degree distribution of an actual network with a Poisson distribution, we can evaluate whether the network has random connectivity.  

2. **Network Randomness**  
・ A comparison with the Poisson distribution helps determine whether a network was generated randomly or according to some rule. For instance, if the degree distribution deviates significantly from the Poisson distribution, the network may have **scale-free** properties.  

3. **Model Fit**  
・ The comparison with the Poisson distribution is a metric to evaluate how well a network model reflects real-world networks. If the degree distribution closely matches the Poisson distribution, the network can be well-represented by the Erdős-Rényi random graph model.

## Script Explanation  
This script aims to compare Erdős-Rényi graphs (random graphs) with the Poisson distribution. Specifically, it demonstrates that the degree distribution of Erdős-Rényi graphs generated with different probabilities $p$ follows the Poisson distribution.

1. **Generating Random Graphs with Different Probabilities**  
・ The script generates Erdős-Rényi graphs with different probabilities.  
・ The function `nx.erdos_renyi_graph(n, p)` takes the number of nodes 𝑛 and the probability 𝑝 of edge creation as parameters and generates a random graph where each edge exists with probability 𝑝.  
・ For example, `er01 = nx.erdos_renyi_graph(100, 0.01)` generates an Erdős-Rényi graph with 100 nodes, where each edge has a 0.01 probability of existing.

2. **Graph Visualization and Degree Distribution Plotting**  
・ The script visualizes the generated random graph and plots its degree distribution.  
・ Using `nx.draw` and `plt.plot(nx.degree_histogram(graph))`, it displays the graph and visualizes the degree distribution.  
・ The degree distribution shows how many connections (edges) each node in the graph has, helping to understand the structure of the random graph.

3. **Comparison with the Poisson Distribution**  
・ In the latter part of the script, a Poisson distribution is generated and compared with the degree distribution of the random graph.  
・ The functions `p1(x), p3(x), p5(x), p10(x)` generate Poisson distributions with means of 𝜆 = 1, 3, 5, and 10, respectively.  
・ By comparing the Poisson distribution with the degree distribution of the random graph, the script examines the similarities and differences between the network properties of the random graph and the theoretical distribution of Poisson.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import scipy.special

# Define Poisson distribution
def p1(x):
    return (1**x)*(np.e**(-1))/scipy.special.factorial(x)

def p3(x):
    return (3**x)*(np.e**(-3))/scipy.special.factorial(x)

def p5(x): 
    return (5**x)*(np.e**(-5))/scipy.special.factorial(x)

def p10(x):
    return (10**x)*(np.e**(-10))/scipy.special.factorial(x)

# Set a large figure size for the entire plot
plt.figure(figsize=(20, 15))

# Generate Erdős-Rényi graphs with different probabilities (number of nodes set to 50)
er01 = nx.erdos_renyi_graph(50, 0.01)
plt.subplot(3, 4, 1)
nx.draw(er01, node_size=50, node_color='red')
plt.title("ER (p=0.01)")

plt.subplot(3, 4, 5)
plt.plot(nx.degree_histogram(er01))
plt.title("Degree Histogram")

er1 = nx.erdos_renyi_graph(50, 0.03)
plt.subplot(3, 4, 2)
nx.draw(er1, node_size=50, node_color='red')
plt.title("ER (p=0.03)")

plt.subplot(3, 4, 6)
plt.plot(nx.degree_histogram(er1))
plt.title("Degree Histogram")

er5 = nx.erdos_renyi_graph(50, 0.05)
plt.subplot(3, 4, 3)
nx.draw(er5, node_size=50, node_color='red')
plt.title("ER (p=0.05)")

plt.subplot(3, 4, 7)
plt.plot(nx.degree_histogram(er5))
plt.title("Degree Histogram")

er8 = nx.erdos_renyi_graph(50, 0.1)
plt.subplot(3, 4, 4)
nx.draw(er8, node_size=50, node_color='red')
plt.title("ER (p=0.1)")

plt.subplot(3, 4, 8)
plt.plot(nx.degree_histogram(er8))
plt.title("Degree Histogram")

# Plot Poisson distributions
plt.subplot(3, 4, 9)
x = np.arange(0, 4, 1)
y = p1(x)
plt.plot(x, y)
plt.title("Poisson (λ=1)")

plt.subplot(3, 4, 10)
x = np.arange(0, 10, 1)
y = p3(x)
plt.plot(x, y)
plt.title("Poisson (λ=3)")

plt.subplot(3, 4, 11)
x = np.arange(0, 12, 1)
y = p5(x)
plt.plot(x, y)
plt.title("Poisson (λ=5)")

plt.subplot(3, 4, 12)
x = np.arange(0, 20, 1)
y = p10(x)
plt.plot(x, y)
plt.title("Poisson (λ=10)")

plt.tight_layout()
plt.show()
```

## Configuration Model  
The configuration model is a method used to generate random graphs with a specific degree distribution (the distribution of the number of edges each node has). This model is widely used to replicate the degree distribution of real-world networks.

### Generation Process  
**Degree Assignment:** First, the degree (number of connections) for each node is specified. For example, Node A might have 3 edges, while Node B has 2 edges.  

**Edge Assignment:** During the edge assignment process, self-loops (edges that connect a node back to itself) and multiedges (multiple edges between the same pair of nodes) can occur. In the context of friendships, a self-loop would represent "counting oneself as a friend." Since self-loops and multiedges are unnatural in real-world networks, these are adjusted to create more realistic networks.  

**Adjustment of Self-Loops and Multiedges:** Self-loops and multiedges may arise during the edge assignment process. Adjusting these ensures the generation of more realistic networks.

### Configuration Model Calculation  
As a concrete example, consider a network with 10 nodes, where each node has the following degrees:  
Node 1: 3  
Node 2: 2  
Node 3: 1  
... (remaining nodes omitted)  
First, assign each node the number of "half-edges" corresponding to its degree, then randomly pair these half-edges to form complete edges. By repeating this process, a random graph with the specified degree distribution is created.

### Applications of the Configuration Model  
**Social Network Analysis:** It is used to replicate the structure of real-world social networks and analyze friendship patterns.  
**Infrastructure Network Design:** The model helps simulate networks with specific degree distributions in fields like power grids or communication networks, aiding in efficient design.  
**Biological Network Research:** It is employed to replicate the structure of genetic or neural networks, helping researchers understand their functions and characteristics.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the degree sequence
deg_seq = [0, 1, 1, 2, 2, 2, 2, 3, 3, 4]

# Create the first plot (1st of 3 sections in total)
plt.subplot(1, 3, 1)
# Generate a graph using the configuration model
G1 = nx.configuration_model(deg_seq)
# Draw the graph using the spring layout
nx.draw_spring(G1, node_size=10, node_color='red')

# Create the second plot (2nd of 3 sections in total)
plt.subplot(1, 3, 2)
# Generate a graph using the configuration model
G2 = nx.configuration_model(deg_seq)
# Draw the graph using the spring layout
nx.draw_spring(G2, node_size=10, node_color='red')

# Create the third plot (3rd of 3 sections in total)
plt.subplot(1, 3, 3)
# Generate a graph using the configuration model
G3 = nx.configuration_model(deg_seq)
# Draw the graph using the spring layout
nx.draw_spring(G3, node_size=10, node_color='red')

# Display all plots
plt.show()
```

## Scale-Free Networks  
A **scale-free network** is a network in which the number of connections (degree) among nodes follows a specific pattern, characterized by the presence of nodes called <span style="color: red;">**hubs**</span>, which have a significantly higher number of connections. In general, scale-free networks tend to have a **degree distribution that follows a power law**.

### Key Features  
- Degree distribution follows a power law  
- Coexistence of a few high-degree nodes and many low-degree nodes  
- High efficiency in connections and fast information propagation  
- Self-organization as the network grows  

For example, in social media platforms, some users have a large number of followers and wield more influence than others. These users, often called influencers, function as hubs in the network. As a result, the **distribution of followers follows a power law, where a small number of users have a very large number of followers, while the majority of users have relatively few followers.** Information propagation also centers around these influencers and high-follower users, allowing posts to spread quickly across many users on social media.

Scale-free networks are observed in various natural phenomena and human-made systems, including social networks, the topology of the internet, and protein interaction networks.

### Scale-Free Nature in Social Networks  
Social network services (SNS) are typical examples of scale-free networks.  
- **Role of Influencers**: Some users (influencers) have a large number of followers and act as <span style="color: red;">**hubs**</span> in the network.  
- **Follower Distribution**: The distribution of followers follows a power law, where a small number of users have a vast number of followers, while the majority have relatively few.  
- **Characteristics of Information Diffusion**: Information spreads rapidly, centered around hub users, with posts propagating widely in a short period of time.

As shown in these examples, understanding and analyzing scale-free networks is essential for comprehending various network phenomena in modern society and developing strategies to address them.

---
In the following script, you can observe that the degree distribution of a scale-free network follows a power law, which is a key characteristic of such networks. This distribution highlights the phenomenon where a few hubs have many links, while most nodes have relatively few links.

To better understand the features of scale-free networks, the script includes the addition of a cumulative degree distribution plot and a degree-rank plot:  
**- Cumulative Degree Distribution Plot**  
A distinguishing feature of scale-free networks is that their degree distribution follows a power law. By plotting the cumulative degree distribution, this feature can be visually confirmed. The cumulative distribution shows how the number of nodes with a degree greater than or equal to a specific value decreases.  
**- Degree and Node Rank Plot**  
Plotting the degree and node rank on a logarithmic scale further emphasizes the scale-free nature of the network. By using rank on the horizontal axis and degree on the vertical axis, this plot illustrates the variability in node popularity.

<span style="color: red;">**Note!**</span>: **To adjust the cumulative degree distribution (CCDF) plot so that it appears linear (follows a power law), it is effective to modify the size and structure of the generated scale-free network, including more nodes and edges. (For example, 1000 nodes, 3 edges)**

The node with the highest degree is highlighted in red.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Generate a scale-free network
# (Adjust the number of nodes and edges to make the CCDF linear)
ba_graph = nx.barabasi_albert_graph(100, 3)

# Identify the node with the most edges
max_degree_node = max(ba_graph.nodes(), key=lambda n: ba_graph.degree(n))

# Calculate the number of nodes, edges, and average degree
num_nodes = nx.number_of_nodes(ba_graph)
num_edges = nx.number_of_edges(ba_graph)
average_degree = 2 * num_edges / num_nodes

# Output the results to the console
print("Number of nodes:", num_nodes)
print("Number of edges:", num_edges)
print("Average degree:", '{:.3f}'.format(average_degree))

# Visualize the network
plt.figure(figsize=(18, 12))

# List to define node colors
node_colors = ['red' if n == max_degree_node else 'blue' for n in ba_graph.nodes()]

plt.subplot(221)
nx.draw(ba_graph, with_labels=False, node_color=node_colors, node_size=30, edge_color='gray')
plt.title('Scale-Free Network with Node with Maximum Degree Highlighted')

# Calculate and plot the degree distribution (log scale)
degree_sequence = sorted([d for n, d in ba_graph.degree()], reverse=True)

plt.subplot(222)
plt.hist(degree_sequence, bins=np.arange(min(degree_sequence) - 0.5, max(degree_sequence) + 1.5, 1), alpha=0.7, log=True)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution (Log Scale)')

# Plot the cumulative degree distribution
degree_count = np.bincount(degree_sequence)
cdf = np.cumsum(degree_count) / sum(degree_count)

plt.subplot(223)
plt.plot(np.arange(len(cdf)), 1 - cdf, marker='o', linestyle='-', color='blue')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Degree')
plt.ylabel('1 - CDF')
plt.title('Complementary Cumulative Degree Distribution (CCDF)')

# Plot the degree-rank distribution
ranks = np.arange(1, len(degree_sequence) + 1)
plt.subplot(224)
plt.plot(ranks, degree_sequence, marker='o', linestyle='-', color='green')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Rank')
plt.ylabel('Degree')
plt.title('Degree vs. Rank (Log-Log Scale)')

plt.tight_layout()
plt.show()
```

## Small-World Networks  
**Small-world networks are characterized by relatively short average path lengths, where connections between nodes are made with only a few hops, and high clustering coefficients, indicating the presence of densely connected regions (clusters) within the network.** This feature places small-world networks between random and regular graphs and is commonly generated using models like the Watts-Strogatz model.

In small-world networks, individual elements within the network are directly connected and form clusters, similar to groups of friends or colleagues in real-world social networks. For example, in social networks where a friend's friend is often directly connected to you, information or influence spreads over short paths. Additionally, since some central figures are directly connected to many people, the structure facilitates the efficient spread of information.

<span style="color: red;">**The main difference between scale-free networks and small-world networks lies in their degree distribution.**</span> In scale-free networks, the degree distribution of nodes follows a power law, where most nodes have few connections and a few **hubs** (high-degree nodes) have many connections.  
On the other hand, in small-world networks, **there is no specific rule for the degree distribution**, and they fall somewhere between randomness and regularity, showing a non-uniform distribution.

### Differences Between Scale-Free and Small-World Networks  
1. **Degree Distribution**  
・ **Scale-Free:** The degree distribution of nodes follows a **power law**, meaning most nodes have few connections, while a few **hubs** (high-degree nodes) have many connections.  
・ **Small-World:** The degree distribution has **no specific rule**, and the network is positioned between randomness and regularity, with no particular pattern required in the distribution.

2. **Network Generation Mechanism**  
・ **Scale-Free:** Generated based on a **preferential attachment** mechanism, where the probability of connecting to existing nodes is proportional to their degree (e.g., BA model).  
・ **Small-World:** Generated by **randomly rewiring** a regular structure (e.g., Watts-Strogatz model).

3. **Information Propagation Characteristics**  
・ **Scale-Free:** Information is **efficiently propagated through hub nodes**, leading to a strong influence on the entire network.  
・ **Small-World:** Information spreads easily over **short paths** and tends to be concentrated in **clustered structures**, allowing for focused dissemination.

## A Side Note...  
The GN (Girvan-Newman) benchmark is different from the Girvan-Newman algorithm.  

**Girvan-Newman Algorithm**  
This is an algorithm used for community detection in networks. It analyzes the structure of the network using edge betweenness centrality and, based on the results, identifies communities (modules). Edge betweenness centrality is a measure that indicates how much each edge contributes to the shortest paths between different pairs of nodes in the network.

**GN (Girvan-Newman) Benchmark**  
This benchmark is used in network science and computer science to evaluate the performance of network algorithms. The GN benchmark generates artificial networks with specific properties (e.g., known community structure) to test the efficiency and accuracy of algorithms. While named after Girvan and Newman, it is not directly related to the specific algorithm.

In short, the Girvan-Newman algorithm refers to a specific network analysis method, while the GN benchmark refers to a method for generating artificial networks to evaluate or compare algorithms.

## Generation and Visualization of LFR Benchmark Graphs  
The LFR (Lancichinetti-Fortunato-Radicchi) benchmark is a model designed to reflect the characteristics of real-world networks, generating networks with heavy-tailed (power-law) degree distributions and community size distributions. Due to this feature, hub nodes (high-degree nodes) and large communities are more likely to appear, faithfully reproducing the complexity of real-world networks.

By adjusting the mixing parameter, the LFR benchmark allows control over the modularity (cohesiveness of communities) of the generated network by regulating the probability of edges between communities. In this sense, the LFR benchmark provides a more realistic model compared to the simpler GN (Girvan-Newman) benchmark. Specifically, it reproduces important features observed in real data, such as heavy-tailed degree distributions and uneven community size distributions.

The LFR benchmark is widely used to evaluate new network analysis methods and community detection algorithms. By comparing the characteristics of the generated networks with real-world data, the validity and applicability of the model can be assessed.

By running this script, you will understand the process of generating an LFR benchmark graph and visually grasp the structure of the generated network.

1. **Generation of the Benchmark Graph**  
   The `LFR_benchmark_graph` function is used to generate an LFR benchmark graph based on specified parameters (e.g., number of nodes, exponent of the degree distribution, **exponent of the community size distribution**, mixing parameter, etc.). This graph serves as an ideal model for mimicking complex real-world networks.  

2. **Community Extraction**  
   Communities to which each node belongs are extracted from the generated network and represented as a list of integers. Each node belongs to a unique community, providing a clear understanding of the community structure.

3. **Node Size Setting**  
   The size of each node is determined based on its degree. This script allows for adjusting the scaling factor of the node size, enabling the emphasis of important nodes visually (e.g., using a 10x scaling factor).

4. **Network Visualization**  
   Nodes are positioned using the spring layout, and each node is displayed in a color corresponding to its community. The `plt.cm.jet` colormap is used for coloring, making it easy to intuitively understand the overall structure of the network.

```python
import networkx as nx
from networkx.generators.community import LFR_benchmark_graph
import matplotlib.pyplot as plt
import numpy as np
import powerlaw  # Import the powerlaw library

# Increase the number of nodes to approximate a power law
n = 1200  # Set the number of nodes to 1200
tau1 = 3  # Exponent for the degree distribution of nodes
tau2 = 1.5  # Exponent for the community size distribution
mu = 0.1  # Mixing parameter

# Adjust min_community and max_community
min_community = 50  # Set the minimum community size to 50
max_community = 100  # Set the maximum community size to 100

# Generate the LFR network
G = LFR_benchmark_graph(n, tau1, tau2, mu, average_degree=10, min_community=min_community, max_community=max_community, seed=42, max_iters=500)

# Convert the community of each node to an integer list
communities = {n: list(d['community'])[0] for n, d in G.nodes(data=True)}

# Set node sizes based on node degrees
node_sizes = [G.degree(n) * 10 for n in G.nodes()]  # 10 is the scaling factor. Adjust as needed.

# Set up three side-by-side subplots
plt.figure(figsize=(24, 6))  # Create a wide figure

# Plot the LFR network
plt.subplot(141)  # First subplot in a 1x4 layout
pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=node_sizes, node_color=[communities[n] for n in G.nodes()], cmap=plt.cm.jet)
plt.title("LFR Benchmark Graph")

# Calculate and plot the degree distribution (log scale)
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
plt.subplot(142)  # Second subplot in a 1x4 layout
plt.hist(degree_sequence, bins=np.arange(min(degree_sequence) - 0.5, max(degree_sequence) + 1.5, 1), alpha=0.7, log=True)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution (Log Scale)')

# Plot the cumulative degree distribution
degree_count = np.bincount(degree_sequence)
cdf = np.cumsum(degree_count) / sum(degree_count)
plt.subplot(143)  # Third subplot in a 1x4 layout
plt.plot(np.arange(len(cdf)), 1 - cdf, marker='o', linestyle='-', color='blue')
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Degree')
plt.ylabel('1 - CDF')
plt.title('Complementary Cumulative Degree Distribution (CCDF)')

# Power-law fitting and test
plt.subplot(144)  # Fourth subplot in a 1x4 layout
fit = powerlaw.Fit(degree_sequence)
fit.power_law.plot_pdf(color='b', linestyle='--', label='Power law fit')
fit.plot_pdf(color='r', linestyle='-', label='Empirical data')

# Fitting to the power law and comparison with other distributions
R, p = fit.distribution_compare('power_law', 'exponential', normalized_ratio=True)
plt.title(f'Power Law Fit\nR={R:.2f}, p={p:.2f}')
plt.legend()

plt.tight_layout()
plt.show()
```

## Comparison of Degree Distributions Between the Karate Club Graph and the Barabási-Albert Model, and the Kolmogorov-Smirnov Test

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import collections
import scipy.stats as stats

# Load the Karate Club graph
G = nx.karate_club_graph()

# Get the number of nodes and edges
n = G.number_of_nodes()
m = G.number_of_edges()

# Get the degree distribution of the Karate Club graph
degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
degree_count = collections.Counter(degree_sequence)
deg, cnt = zip(*degree_count.items())

# Plot the degree distribution to check for scale-free properties
plt.figure(figsize=(12, 6))

# Adjust the starting point of the x-axis to the minimum degree
plt.subplot(121)
plt.bar(deg, cnt, width=0.80, color='b', align='center')
plt.title("Degree Histogram (Karate Club)")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.xlim(min(deg) - 0.5, max(deg) + 0.5)

plt.subplot(122)
plt.loglog(deg, cnt, marker='o', linestyle='None')
plt.title("Log-Log Degree Distribution (Karate Club)")
plt.ylabel("Frequency")
plt.xlabel("Degree")

plt.tight_layout()
plt.show()

# Generate a scale-free graph using the Barabási-Albert model
mean_degree = np.mean(degree_sequence)
m = int(mean_degree / 2)  # The parameter m in the BA model is the number of existing nodes each new node connects to

G_ba = nx.barabasi_albert_graph(n, m)

# Get the degree distribution of the BA model
degree_sequence_ba = sorted([d for n, d in G_ba.degree()], reverse=True)
degree_count_ba = collections.Counter(degree_sequence_ba)
deg_ba, cnt_ba = zip(*degree_count_ba.items())

# Plot the degree distribution for comparison between the BA model and Karate Club graph
plt.figure(figsize=(12, 6))

# Adjust the starting point of the x-axis to the minimum degree
plt.subplot(121)
plt.bar(deg_ba, cnt_ba, width=0.80, color='g', align='center')
plt.title("Degree Histogram (BA Model)")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.xlim(min(deg_ba) - 0.5, max(deg_ba) + 0.5)

plt.subplot(122)
plt.loglog(deg_ba, cnt_ba, marker='o', linestyle='None')
plt.title("Log-Log Degree Distribution (BA Model)")
plt.ylabel("Frequency")
plt.xlabel("Degree")

plt.tight_layout()
plt.show()

# Visualization of the Karate Club graph and the BA model graph
plt.figure(figsize=(12, 6))

plt.subplot(121)
nx.draw_spring(G, with_labels=True, node_color='skyblue', edge_color='gray', node_size=500, font_size=10)
plt.title("Original Karate Club Graph")

plt.subplot(122)
nx.draw_spring(G_ba, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
plt.title("BA Model Graph")

plt.tight_layout()
plt.show()

# Quantitative evaluation
mean_degree_ba = np.mean(degree_sequence_ba)
max_degree_karate = max(degree_sequence)
max_degree_ba = max(degree_sequence_ba)

print("Karate Club Graph:")
print(f"Mean Degree: {mean_degree:.2f}")
print(f"Max Degree: {max_degree_karate}")

# Display the number of neighbors connected to the node with the maximum degree in the Karate Club graph
for node in G.nodes():
    if G.degree(node) == max_degree_karate:
        neighbors_count = len(list(G.neighbors(node)))
        print(f"Nodes with max degree ({max_degree_karate}) in the Karate Club graph are connected to {neighbors_count} neighbors.")

print("\nBA Model Graph:")
print(f"Mean Degree: {mean_degree_ba:.2f}")
print(f"Max Degree: {max_degree_ba}")

# Statistical test (Kolmogorov-Smirnov test)
ks_stat, p_value = stats.ks_2samp(degree_sequence, degree_sequence_ba)

print("\nKolmogorov-Smirnov Test:")
print(f"KS Statistic: {ks_stat:.2f}")
print(f"P-value: {p_value:.2f}")

# Interpretation of the test result
if p_value > 0.05:
    print("The degree distributions are not significantly different (fail to reject H0).")
else:
    print("The degree distributions are significantly different (reject H0).")
```

### The Meaning of Running the KS Test  
The KS test is used to determine whether two distributions are statistically significantly different. In this case, it compares the degree distributions of the Karate Club Graph and the BA Model Graph.

### Significance and Purpose  
**Evaluating Model Validity:**  
The Karate Club Graph is based on real-world data, while the BA Model Graph is a randomly generated graph using the Barabási-Albert model.  
By comparing the degree distributions of these two graphs with the KS test, we can evaluate how well the BA model replicates real-world data (the Karate Club Graph).

**Statistical Verification:**  
Basic statistics like mean and maximum degree can only provide limited insights into the overall shape of the distribution. The KS test evaluates the differences between the cumulative distribution functions, allowing for a more detailed comparison.

**Testing the Null Hypothesis:**  
The null hypothesis (H0) is that the two samples come from the same distribution. This test helps determine whether the structures of the two graphs are statistically equivalent.

### Conclusion  
The KS test results in a high p-value of 0.86, meaning we fail to reject the null hypothesis. In other words, the degree distributions of the Karate Club Graph and the BA Model Graph are not statistically significantly different.

- It suggests that the BA model is able to replicate the degree distribution of the Karate Club Graph.  
- It provides guidance for model selection and parameter adjustment.  
- It helps deepen our understanding of network structures by comparing real-world networks with random models.
