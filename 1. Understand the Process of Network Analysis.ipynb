## Basic Concepts of Networks

Refers to the individual points within a network. For example, in a social network, nodes represent individuals.

Refers to the lines connecting nodes. These represent relationships or interactions between nodes.

Refers to the entire graph, which is composed of the nodes and edges.

## Calculation of Structural Features of Networks

Structural features of networks are important metrics for quantitatively understanding the scale, shape, and relationships between nodes in a network. By calculating these features, we can gain insights into the nature of the network, which can be useful for further analysis.

1. **Overall Structure of the Network**  
   The most basic features include the number of nodes (vertices) and the number of edges. These indicate the scale of the network and serve as a starting point for understanding how many elements are connected. Additionally, **density** refers to the proportion of actual edges out of all possible edges, which evaluates how "dense" the network is.

2. **Evaluating Node Importance and Influence**  
   The **degree** of a node, which shows how many edges it has, is an indicator of how connected that node is to other nodes. Furthermore, to identify nodes that play important roles within the network, various **centrality measures** (such as degree centrality, betweenness centrality, and eigenvector centrality) are used. These measures help pinpoint nodes that act as information hubs or control the flow within the network.

3. **Understanding Local Patterns**  
   To analyze how the network is structured in finer detail, metrics like the **clustering coefficient** and **connected components** are calculated. The clustering coefficient shows how tightly a node’s neighbors are interconnected, which is useful for understanding the formation of groups or subgroups within the network. Connected components, on the other hand, represent sets of nodes that can reach each other, revealing independent communities within the network.

4. **Network Efficiency and Robustness**  
   Metrics like **average path length** and **diameter** reflect the overall efficiency of the network. The average path length is the mean of the shortest paths between any two nodes, indicating how efficiently information spreads. The diameter, which is the longest shortest path between any two nodes, shows the maximum extent of the network. These metrics are crucial for assessing how effectively a network transmits information and how resilient it is to failure.

5. **Comparison and Model Evaluation**  
   When comparing different networks, these structural features help analyze the strengths and weaknesses of each network. Additionally, when comparing real networks with theoretical models (such as random graphs or small-world models), the calculation of structural features helps evaluate the model's fit and identify areas for improvement.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Function to display additional labels
def add_features_to_plot(G, pos, ax, features, vertical=True):
    labels = {node: f"{features[node]}" for node in G.nodes()}
    for node, (x, y) in pos.items():
        if vertical:
            ax.text(x, y + 0.1, labels[node], fontsize=12, ha='center', va='bottom', color='blue', fontweight='bold')
        else:
            ax.text(x + 0.1, y, labels[node], fontsize=12, ha='left', va='center', color='blue', fontweight='bold')

# Specify plot size
plt.figure(figsize=(15, 15))

# Generate a random graph with 10 vertices and edges with probability 0.3, and draw it using a circular layout.
rnd = nx.gnp_random_graph(10, 0.3)
pos = nx.circular_layout(rnd)
plt.subplot(331)
nx.draw_networkx_nodes(rnd, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(rnd, pos)
nx.draw_networkx_labels(rnd, pos, font_weight='bold')
plt.title('Random Graph (10 nodes, p=0.3)')
add_features_to_plot(rnd, pos, plt.gca(), nx.degree(rnd), vertical=True)

# Draw the Petersen graph using a shell layout.
petersen = nx.petersen_graph()
pos = nx.shell_layout(petersen, nlist=[range(5, 10), range(5)])
plt.subplot(332)
nx.draw_networkx_nodes(petersen, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(petersen, pos)
nx.draw_networkx_labels(petersen, pos, font_weight='bold')
plt.title('Petersen Graph')
add_features_to_plot(petersen, pos, plt.gca(), nx.degree(petersen), vertical=True)

# Draw a cycle graph with 10 vertices using a circular layout.
cycle = nx.cycle_graph(10)
pos = nx.circular_layout(cycle)
plt.subplot(333)
nx.draw_networkx_nodes(cycle, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(cycle, pos)
nx.draw_networkx_labels(cycle, pos, font_weight='bold')
plt.title('Cycle Graph (10 nodes)')
add_features_to_plot(cycle, pos, plt.gca(), nx.degree(cycle), vertical=True)

# Draw a complete graph with 10 vertices using a circular layout.
K_10 = nx.complete_graph(10)
pos = nx.circular_layout(K_10)
plt.subplot(334)
nx.draw_networkx_nodes(K_10, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(K_10, pos)
nx.draw_networkx_labels(K_10, pos, font_weight='bold')
plt.title('Complete Graph (10 nodes)')
add_features_to_plot(K_10, pos, plt.gca(), nx.degree(K_10), vertical=True)

# Draw a complete bipartite graph with 5 vertices on each side using a circular layout.
K_5_5 = nx.complete_bipartite_graph(5, 5)
pos = nx.circular_layout(K_5_5)
plt.subplot(335)
nx.draw_networkx_nodes(K_5_5, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(K_5_5, pos)
nx.draw_networkx_labels(K_5_5, pos, font_weight='bold')
plt.title('Complete Bipartite Graph (5, 5)')
add_features_to_plot(K_5_5, pos, plt.gca(), nx.degree(K_5_5), vertical=True)

# Draw a barbell graph using a spring layout.
barbell = nx.barbell_graph(5, 5)
pos = nx.spring_layout(barbell)
plt.subplot(336)
nx.draw_networkx_nodes(barbell, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(barbell, pos)
nx.draw_networkx_labels(barbell, pos, font_weight='bold')
plt.title('Barbell Graph (m1=5, m2=5)')
add_features_to_plot(barbell, pos, plt.gca(), nx.degree(barbell), vertical=False)

# Draw a star graph with 1 central node and 9 peripheral nodes using a spring layout.
star = nx.star_graph(9)
pos = nx.spring_layout(star)
plt.subplot(337)
nx.draw_networkx_nodes(star, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(star, pos)
nx.draw_networkx_labels(star, pos, font_weight='bold')
plt.title('Star Graph (9 peripheral nodes)')
add_features_to_plot(star, pos, plt.gca(), nx.degree(star), vertical=False)

# Draw a wheel graph with 10 vertices using a spring layout.
wheel = nx.wheel_graph(10)
pos = nx.spring_layout(wheel)
plt.subplot(338)
nx.draw_networkx_nodes(wheel, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(wheel, pos)
nx.draw_networkx_labels(wheel, pos, font_weight='bold')
plt.title('Wheel Graph (10 nodes)')
add_features_to_plot(wheel, pos, plt.gca(), nx.degree(wheel), vertical=False)

# Draw a graph based on the Barabási-Albert model using a spring layout.
ba = nx.barabasi_albert_graph(10, 2)
pos = nx.spring_layout(ba)
plt.subplot(339)
nx.draw_networkx_nodes(ba, pos, node_size=200, node_color='red')
nx.draw_networkx_edges(ba, pos)
nx.draw_networkx_labels(ba, pos, font_weight='bold')
plt.title('Barabási-Albert Graph (n=10, m=2)')
add_features_to_plot(ba, pos, plt.gca(), nx.degree(ba), vertical=False)

plt.tight_layout(pad=3.0)
plt.show()
```

## Calculation of Network Density and Clustering Coefficient

Network density and the clustering coefficient are important metrics that describe the structure of a network.

Network density represents the ratio of the number of actual edges to the number of all possible edges. A higher density indicates that the nodes are more tightly connected, allowing information or influence to spread more quickly. Using the number of nodes |V| and the number of edges |E| in the network, the density D is calculated by the following formula:

$D = \frac{2|E|}{|V|(|V| - 1)}$

On the other hand, the clustering coefficient is another key metric that describes the structure of a network. It measures the ratio of the number of actual edges between a node and its neighboring nodes to the maximum number of possible edges between them. The clustering coefficient ranges from 0 to 1, where a higher value indicates that the node and its surroundings are more closely connected.

The clustering coefficient provides insights into the nature of the network. For example, when the clustering coefficient is close to 0, the network tends to be randomly structured. If it is close to 1, the network exhibits a high level of clustering. In real-world networks, social networks typically have high clustering coefficients, while power grid networks tend to have low clustering coefficients.

In this script, we use NetworkX's clustering coefficient calculation function to display the average clustering coefficient for each graph on the plot. This allows us to compare how clustered different graph structures, such as random graphs and network models, are.

The clustering coefficient of a node \( i \), denoted as \( C_i \), is calculated using the number of actual edges \( e_i \) between the neighboring nodes of that node and its degree (number of connections) \( k_i \), according to the following formula:

$C_i = \frac{2 e_i}{k_i (k_i - 1)}$

These metrics are widely used to quantify the structure and characteristics of networks and to compare and analyze different types of networks.

## Centrality Calculation

Centrality calculation is a method used to find **nodes that play particularly important roles** within a network. These important roles refer to positions or functions that greatly influence the overall structure and function of the network, such as facilitating the rapid transmission of information or connecting different parts of the network. By using centrality metrics, it is possible to quantitatively evaluate and understand the significance of specific nodes in the network.

Here are some representative centrality metrics and their meanings:

1. **Degree Centrality**  
   - Indicates the **number of other nodes directly connected** to a given node. Nodes with a high degree of centrality are directly connected to many other nodes, meaning they can have a significant impact on information flow and network structure.

2. **Eigenvector Centrality**  
   - Evaluates the importance of a node by considering the **importance of the nodes it is connected to**. This means that nodes connected to other important nodes will have high centrality, not just those with many connections.

3. **Closeness Centrality**  
   - Measures the **average shortest path length** from a given node to all other nodes. Nodes with high closeness centrality can reach all other nodes in the network quickly, making them efficient in terms of information dissemination.

4. **Betweenness Centrality**  
   - Measures how often a node appears on the shortest paths between other nodes. Nodes with high betweenness centrality serve as **bridges** (hubs) connecting different parts of the network, and they play a crucial role in maintaining network cohesion and information flow.

### The Significance of Centrality Calculation

This script calculates the degree of each node in the Karate Club dataset and visualizes the network's connectivity (centrality) based on several centrality definitions. By visualizing how the graph changes with different centrality metrics, you can intuitively understand the network's structure.

Centrality calculations are useful in various real-world applications, such as:

- **Social Network Analysis**: Identifying key individuals (influencers).  
- **Internet Traffic Management**: Identifying major hubs and bottlenecks to optimize data transfer.  
- **Pathogen Spread Prediction**: Pinpointing critical points to prevent the spread of diseases.  
- **Organizational Management**: Identifying key contact points and understanding information flow within an organization.

By analyzing networks using centrality metrics, you can identify important nodes and gain a deeper understanding of the network’s overall structure and function.  
*Note: The `nx.draw_networkx_nodes` function does not accept a `norm` argument. Therefore, color scale normalization is done through colormap settings.*

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import os

# Define the drawing function
def draw_h(G, pos, measures, measure_name, ax):
    """
    Draws the network graph based on the specified centrality measure.

    Parameters:
    G (networkx.Graph): The network graph to be drawn.
    pos (dict): A dictionary representing the positions of the nodes.
    measures (dict): A dictionary storing the centrality measures of the nodes.
    measure_name (str): The name of the centrality measure to be drawn.
    ax (matplotlib.axes.Axes): The Axes object of the subplot where the graph will be drawn.

    Drawing Procedure:
    1. Draw nodes: The nodes are drawn based on the specified centrality measure.
    2. Set node colors: The colors are normalized using a logarithmic scale.
    3. Draw node labels: Labels are drawn on the nodes.
    4. Draw edges: The edges between the nodes are drawn.
    5. Set title and axis: The title is set for the graph, and the axis is turned off.
    """
    # Normalization
    vmin = min(measures.values())
    vmax = max(measures.values())
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250,
                                  cmap=plt.cm.plasma,
                                  node_color=list(measures.values()),
                                  nodelist=list(measures.keys()),
                                  ax=ax)
    
    edges = nx.draw_networkx_edges(G, pos, ax=ax)
    ax.set_title(measure_name)
    ax.axis('off')
    
    # Add color bar
    sm = plt.cm.ScalarMappable(cmap=plt.cm.plasma, norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax)
    
    nodes = nx.draw_networkx_nodes(G, pos, node_size=250,
                                  cmap=plt.cm.plasma,
                                  node_color=list(measures.values()),
                                  nodelist=list(measures.keys()), ax=ax)
    nodes.set_norm(mcolors.SymLogNorm(linthresh=0.01, linscale=1))
    nx.draw_networkx_labels(G, pos, font_size=10, font_color='black', font_family='sans-serif', ax=ax)
    edges = nx.draw_networkx_edges(G, pos, ax=ax)
    ax.set_title(measure_name)
    ax.axis('off')

# Change the current directory
new_directory = r"D:\Python\NetworkX\lesmis"
os.chdir(new_directory)

# Specify the file path
file_path = r'lesmis.gml'

# Load the file and draw the graph
G = nx.read_gml(file_path)
pos = nx.spring_layout(G)

# Set figure size, calculate various centrality measures, and draw
plt.figure(figsize=(18, 18))
ax1 = plt.subplot(2, 2, 1)
draw_h(G, pos, nx.degree_centrality(G), 'Degree Centrality', ax1)

ax2 = plt.subplot(2, 2, 2)
draw_h(G, pos, nx.eigenvector_centrality(G), 'Eigenvector Centrality', ax2)

ax3 = plt.subplot(2, 2, 3)
draw_h(G, pos, nx.closeness_centrality(G), 'Closeness Centrality', ax3)

ax4 = plt.subplot(2, 2, 4)
draw_h(G, pos, nx.betweenness_centrality(G), 'Betweenness Centrality', ax4)

plt.tight_layout()
plt.show()
```

# Community Detection

Many networks, such as social networks, contain tightly connected groups of vertices, much like friend groups in a social network. These tightly knit sub-networks are referred to as **communities**. For example, in social networking services (SNS), communities correspond to user groups with shared interests or concerns.

In community detection, similar vertices tend to cluster together, often forming a single community. This provides valuable information for systems, such as when displaying posts on the timeline of SNS users with similar interests. Furthermore, identifying central vertices within a group or those located at the boundaries between different groups is crucial for understanding the influence and spread of information.

Consider the network of characters from *Les Misérables*, for example. By identifying groups of characters who are closely connected within the story, we can discover hidden structures and important groups, such as revolutionaries or family relationships. This deepens our understanding of the narrative and helps us grasp key points in the relationships between characters and the progression of the story.

There are various methods for community detection, and the size and number of detected communities may not always be predetermined. Additionally, there are multiple approaches to defining communities.

### Extract Communities Formed by Similar Edges

This script demonstrates the procedure for extracting and visualizing communities within a network. The following steps outline how the community extraction and visualization are performed:

1. **Importing Libraries**  
   At the beginning of the script, the necessary libraries are imported. These include `networkx` (for network analysis), `matplotlib.pyplot` (for graph visualization), `community` (for community detection algorithms), and `pandas` (for data processing).

2. **Loading the GML File**  
   The script reads the network data from a file named `lesmis.gml`. GML is a common format used for describing network structures.

3. **Creating a DataFrame for Node Attributes**  
   A DataFrame is created from the loaded network, storing node attributes. In this case, an attribute called `myvalue` with an initial value of 0 is set for each node, along with their IDs.

4. **Extracting Communities**  
   The **Girvan-Newman algorithm** is used to extract communities (groups) from the network. This algorithm identifies densely connected parts of the network and divides them into communities.

5. **Visualizing the Network**  
   The extracted community numbers are assigned as attributes to each node, and the layout of the nodes is set using `nx.kamada_kawai_layout`, which slightly spreads the distance between nodes. Finally, using `matplotlib`, the nodes are colored based on their community numbers, and the network is displayed with labels.

By running this script, you can extract and visualize the community structure within the network.  
*Note: Adjustments for community extraction and colormap have not been made, so the visualization may not be entirely appropriate!*

```python
# Import the necessary libraries
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community
import pandas as pd

# Path to the GML file
gml_file = r'D:\Python\NetworkX\lesmis\lesmis.gml'

# Load the GML file
G = nx.read_gml(gml_file)

# Create a DataFrame with node attributes
carac = pd.DataFrame({'ID': G.nodes(), 'myvalue': [0] * len(G.nodes())})

# Extract communities using the Girvan-Newman algorithm
communities_generator = community.girvan_newman(G)
# Get the top-level communities
top_level_communities = next(communities_generator)
# Get the next level of communities
next_level_communities = next(communities_generator)

# Set community numbers as node attributes and draw the network
for m in range(len(next_level_communities)):
    for n in next_level_communities[m]:
        carac.loc[carac.ID == n, 'myvalue'] = m

# Set the layout (spread out the distance between nodes)
pos = nx.kamada_kawai_layout(G)

# Draw the network
plt.figure(figsize=(10, 8))
nx.draw(G, pos, node_color=carac['myvalue'], node_size=120, with_labels=True, cmap=plt.cm.gist_ncar)
plt.show()
```
